```python
from spacy.lang.ru import Russian
import spacy
from spacy import displacy
```

**Описание набора данных**

Исходный текст:

```python
text1 = 'Обработка естественного языка — пересечение машинного обучения и математической лингвистики, направленное на изучение методов анализа и синтеза естественного языка. Сегодня NLP применяется во многих сферах, в том числе в голосовых помощниках, автоматических переводах текста и фильтрации текста. Основными тремя направлениями являются: распознавание речи , понимание естественного языка и генерация естественного языка.'
```

## Токенизация

```python
nlp = spacy.load('ru_core_news_sm')
spacy_text1 = nlp(text1)
spacy_text1
```

```python
for t in spacy_text1:
    print(t)
```

# POS-разметка

```python
for token in spacy_text1:
    print('{} - {} - {}'.format(token.text, token.pos_, token.dep_))
```

# Лемматизация

```python
for token in spacy_text1:
      print(token, token.lemma, token.lemma_)
```

# Выделение именованных сущностей (NER)

```python
for ent in spacy_text1.ents:
    print(ent.text, ent.label_)
```

```python
displacy.render(spacy_text1, style='ent', jupyter=True)
```

```python
print(spacy.explain("ORG"))
```

# Разбор предложения

```python
displacy.render(spacy_text1, style='dep', jupyter=True)
```
